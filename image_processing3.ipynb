```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
my_img=cv2.imread("siyahkedi.jpg")
```

```python
type(my_img)
```

```python
my_img.shape
```

```python
#görüntülerde eşikleme

my_img_gray=cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
```

```python
my_img_gray.shape
```

```python
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
```

```python
#Binary Thresh  
#İkili eşikleme , verilen thresh(eşik) değeri baz alınarak bu eşik değerinin altında kalan piksel değerleri 
#yani örneğin 125'ten küçük piksel değeri siyah rengini alacak(0), büyük veya eşit olanlar ise beyaz rengini(255) alacak.

_,my_binary_thresh_img=cv2.threshold(my_img_gray,thresh=125,maxval=255,type=cv2.THRESH_BINARY)
```

```python
_
```

```python
fig=plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("Original Gray İmage")

plt.subplot(1,2,2)
plt.imshow(my_binary_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Binary Thresh İmage")
```

```python
my_img_gray[100,100]     #125'den büyük ise binary thresh olanda beyaz(255) küçükse siyah(0)
```

```python
my_binary_thresh_img[100,100]   #buradaki yer siyah binary thresh yaptıgımız resimde  
```

```python
#Thresh Inverse 
#az önce yapılan eşik değerinin tam tersi 125 den küçük olan beyaz buyuk olan siyah olacak
_, my_binary_inv_thresh_img=cv2.threshold(my_img_gray,thresh=125,maxval=255,type=cv2.THRESH_BINARY_INV)
```

```python
fig=plt.figure(figsize=(10,7))

plt.subplot(1,3,1)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("Original Gray İmage")

plt.subplot(1,3,2)
plt.imshow(my_binary_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Binary Thresh İmage")

plt.subplot(1,3,3)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Binary Inverse Thresh İmage")
```

```python
print(f"Original = {my_img_gray[500,550]}")
print(f"Binary = {my_binary_thresh_img[500,550]}")
print(f"Inverse = {my_binary_inv_thresh_img[500,550]}")
```

```python
#Thesh Trunc
#beyazlıkları gideriyor gibi
_, my_trunc_thresh_img=cv2.threshold(my_img_gray,thresh=125,maxval=255,type=cv2.THRESH_TRUNC)
```

```python
fig=plt.figure(figsize=(10,7))

plt.subplot(2,2,1)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("Original Gray İmage")

plt.subplot(2,2,2)
plt.imshow(my_binary_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Binary Thresh İmage")

plt.subplot(2,2,3)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Binary Inverse Thresh İmage")

plt.subplot(2,2,4)
plt.imshow(my_trunc_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Trunc Thresh İmage")
```

```python
# Binary Thresholding: belirlenen eşik değerine göre (thresh value),
# orijinal görüntünün pikseli, bu eşik değerinden küçükse siyah rengine
# dönüştürülür yani 0 değerini alır, ilgili piksel eşik değerinden büyükse
# beyaz rengine dönüştürülür yani 255 değerini alır.

# Binary Inverse (Binary_inv) Thresholding: Binary thresholding'in tam tersidir.
# ilgili piksel, eşik değerinden küçükse 255 değerini alır (beyaz rengine
# dönüştürülür), eşik değerinden büyükse 0 değerini alır (siyah rengine
# dönüştürülür.

# Trunc Thresholding: Görüntünün parlak renklerinin kısılması için kullanılır.
# Şöyle ki, ilgili piksel değeri eşik değerinden büyükse eşik değerine eşitlenir
# eşik değerinden küçükse aynı bırakılır.

# TOZERO Thresholding: eğer ilgili piksel değeri, eşik değerinden büyükse, olduğu
# gibi korunur yani değişmez aynı kalır. Yok eğer, piksel değerinden küçükse,
# 0 yapılır yani siyah rengine dönüşür.

# TOZERO Thresholding: TOZERO'nun tam tersidir.
# eğer ilgili piksel değeri, eşik değerinden büyükse, 
# 0 yapılır yani siyah rengine dönüşür.
# Eğer küçükse olduğu gibi korunur yani değişmez aynı kalır.
```

```python
#Resize 

#python'da 2 adet resize() fonksiyonu yazınız bu fonksiyon 1 adet parametre alacak.o da görüntünün kendisi olacak
#görüntüyü büyütürkende küçültürkende genişlik ve yüksekliğin ½25'i kadar olacaktır.
#her iki fonksiyon da geriye,işlenmiş görüntünün kendisini dönürecek.

#cv2.resize() kullanın
#cv2.resize(original_img,(genişlik,yükseklik),cv2.INTER_LANCZOS4)

def kucult(my_img):
    yeni_img=cv2.resize(my_img,(int(my_img.shape[1] * 0.75),int(my_img.shape[0] * 0.75)),cv2.INTER_LANCZOS4)
    return yeni_img

def buyut(my_img):
    yeni_img=cv2.resize(my_img,(int(my_img.shape[1] * 1.25),int(my_img.shape[0] * 1.25)),cv2.INTER_LANCZOS4)
    return yeni_img


fig=plt.figure(figsize=(10,10))

kucuk_img=kucult(my_img_gray)
buyuk_img=buyut(my_img_gray)

plt.subplot(1,3,1)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("Original Gray İmage")

plt.subplot(1,3,2)
plt.imshow(kucuk_img,cmap="gray")
plt.axis("off")
plt.title("Small İmage")


plt.subplot(1,3,3)
plt.imshow(buyuk_img,cmap="gray")
plt.axis("off")
plt.title("Big İmage")

plt.tight_layout()            #etraftaki beyazlıkları kısıyor
plt.savefig("./Image_Resizeng.png",dpi=300)     #netliği korur
```
